```python
import pandas as pd
```

```python
df=pd.read_csv("Dataset1_low.csv")
df.head()
```

```python
target_count=df.class_attribute.value_counts()
print(target_count)

imbalance_target=(target_count[1]/(target_count[0]+target_count[1]))*100
print(imbalance_target)

imbalance_target=(target_count[0]/(target_count[0]+target_count[1]))*100
print(imbalance_target)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
df["class_attribute"].hist()
plt.savefig('dataset1.png')
```

```python
df2=pd.read_csv("breast-cancer.csv",header=None)
df2.head()
#Manually removed 9 rows with instances of NaN values
```

```python
columns= ['Class', 'age', 'menopause', 'tumor-size','inv-nodes', 'node-caps', 'deg-malig', 'breast', 'breast-quad','irradiat']
df2.columns=columns
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2=pd.get_dummies(df2)
df2.head()
```

```python
target_count=df2['Class_no-recurrence-events'].value_counts()
print(target_count)

imbalance_target=(target_count[1]/(target_count[0]+target_count[1]))*100
print(imbalance_target)

imbalance_target=(target_count[0]/(target_count[0]+target_count[1]))*100
print(imbalance_target)
```

```python
df2['Class_no-recurrence-events'].hist()
plt.savefig('dataset2.png')
```

```python
df2.to_csv("Dataset_2_medium.csv")
```
